```python
### Generate paranoia-style message prompts using utils.py

from utils.utils import chat
from utils.prompts import PARANOIA_PROMPT
import json, re, ast

resp = chat(PARANOIA_PROMPT, model="gpt-4.1", max_tokens=3000)
print('Model returned (first 1000 chars):')
print((resp or '')[:1000])

# Robust parsing: try JSON array first, then fall back to extracting JSON objects
messages_list = []
try:
    parsed = json.loads(resp)
    if isinstance(parsed, list):
        messages_list = parsed
except Exception:
    # Find individual {...} objects in the text
    objs = re.findall(r'\{.*?\}', resp or '', flags=re.S)
    for o in objs:
        try:
            messages_list.append(json.loads(o))
        except Exception:
            try:
                messages_list.append(ast.literal_eval(o))
            except Exception:
                continue

print(f'Parsed {len(messages_list)} message objects')

# Save to messages/messages.py (overwrites existing file)
out_path = 'messages/messages.py'
with open(out_path, 'w') as f:
    f.write('messages = [\n')
    for m in messages_list:
        f.write('    ' + json.dumps(m, ensure_ascii=False) + ',\n')
    f.write(']\n')

print(f'Wrote {len(messages_list)} messages to {out_path}')
```
